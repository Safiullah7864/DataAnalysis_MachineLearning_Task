```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
iris = sb.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.shape
```

```python
plt.Figure(figsize=(10,6))
sb.scatterplot(
   data= iris,
   x="petal_length",
   y="petal_width",
   hue="species"
)
plt.title("Petal_lenght vs Petal_width")
plt.show()
```

```python
from numpy import spacing
plt.figure(figsize=(10,6))
plt.hist(
    iris["sepal_length"],
    bins=10,
    rwidth=0.9
)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()

```

```python
plt.figure(figsize=(8,6))

sb.boxplot(
    y=iris["sepal_width"]
)

plt.title("Box Plot of Sepal Width")

plt.show()
```
